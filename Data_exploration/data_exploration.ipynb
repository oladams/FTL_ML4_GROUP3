# Energy Consumption Data Exploration

## 1. Librairies Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress all warnings
warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load Dataset

```python
data = pd.read_csv('./../dataset/KAG_energydata_complete.csv')
data.head(5)
```

## 3. Data Mining or Exploration

### Energy Consumption Over Time

```python
# Convert the 'date' column to datetime type for better manipulation
data['date'] = pd.to_datetime(data['date'])

# Plot the evolution of appliances (Appliances)
plt.figure()
sns.lineplot(data=data, x='date', y='Appliances', label="Appliances")
plt.title("Energy Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("Energy Consumption (Wh)")
plt.legend()
plt.show()
```

### Distribution of Relative Humidity in Room 1

```python
plt.figure()
sns.histplot(data['RH_1'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Relative Humidity in Room 1")
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Frequency")
plt.show()
```

### Correlation Matrix of Environmental Variables

```python
# Select the environmental columns for correlation
env_variables = ['T1', 'RH_1', 'T_out', 'Press_mm_hg', 'Windspeed', 'Visibility', 'Tdewpoint']
correlation_matrix = data[env_variables].corr()

# Display the correlation matrix
plt.figure()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Environmental Variables")
plt.show()
```

### Average Daily Energy Consumption by Day of the Week

```python
# Extract the day of the week
data['day_of_week'] = data['date'].dt.day_name()

# Average consumption per day
daily_avg = data.groupby('day_of_week')['Appliances'].mean()

# Plot the averages
plt.figure()
sns.barplot(x=daily_avg.index, y=daily_avg.values, palette='viridis')
plt.title("Average Daily Energy Consumption by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Energy Consumption (Wh)")
plt.xticks(rotation=45)
plt.show()
```

### Temperature vs Relative Humidity in Room 1

```python
plt.figure()
sns.scatterplot(data=data, x='T1', y='RH_1', alpha=0.6, color="orange")
plt.title("Temperature vs Relative Humidity in Room 1")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Relative Humidity (%)")
plt.show()
```

### Energy Consumption: Appliances vs Lights

```python
plt.figure()
sns.boxplot(data=data[['Appliances', 'lights']], palette="Set2")
plt.title("Energy Consumption: Appliances vs Lights")
plt.ylabel("Energy Consumption (Wh)")
plt.xlabel("Category")
plt.show()
```

### Monthly Average Energy Consumption

```python
# Extract the month and hour to analyze seasonality
data['month'] = data['date'].dt.month
data['hour'] = data['date'].dt.hour

# Average consumption by month
monthly_avg = data.groupby('month')['Appliances'].mean()

plt.figure()
sns.lineplot(x=monthly_avg.index, y=monthly_avg.values, marker="o")
plt.title("Monthly Average Energy Consumption")
plt.xlabel("Month")
plt.ylabel("Energy Consumption (Wh)")
plt.xticks(range(1, 13))
plt.show()
```
